```python
#1st table customer_table
import mysql.connector

class sank:
    def __init__(self, host, user, password, port, database):
        self.host = host
        self.user = user
        self.password = password
        self.port = port
        self.database = database
        self.conn = None
        self.cursor = None

    def connect(self):
        self.conn = mysql.connector.connect(
            host=self.host, user=self.user, password=self.password, port=self.port, database=self.database
        )
        self.cursor = self.conn.cursor()
        print(" Connection successful!")

    def creat_table(self, table_query):
        self.cursor.execute(table_query)
        self.conn.commit()
        print(" Table creation successful!")

    def insert_list_of_tuple(self, insert_query, list_tuple):
        self.cursor.executemany(insert_query, list_tuple)
        self.conn.commit()
        print(" Insertion successful!")

    def fetch_all(self, query):
        """ Fetches all results from a given query """
        self.cursor.execute(query)
        return self.cursor.fetchall()  # Returns a list of tuples
```

```python
MO = """
CREATE TABLE manages_order (
    order_id INT PRIMARY KEY,
    customer_id INT,
    restaurant_id INT,
    order_date DATE,
    delivery_time TIME,
    order_status VARCHAR(255),
    total_amount INT,
    payment_mode VARCHAR(255),
    discount_applied INT,
    feed_back_rating FLOAT,
    FOREIGN KEY (customer_id) REFERENCES Customers_Table(customer_id),
    FOREIGN KEY (restaurant_id) REFERENCES manages_restaurant_infos(restaurant_id)
);

"""

db_obj.creat_table(MO)
```

```python
#10,Find restaurants with an average rating of 4 or above and more than 500 orders:

results = db_obj.fetch_all('''SELECT name, rating, total_orders
FROM manages_restaurant_infos
WHERE rating >= 4 AND total_orders > 500;''')

headers = ["Name", "Rating", "Total Orders"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```

```python
#14,Find the total number of deliveries completed by each delivery person:

results = db_obj.fetch_all('''SELECT dp.name, COUNT(d.delivery_id) AS total_deliveries
FROM delivery_table_data d
JOIN delivery_person_table dp ON d.delivery_person_id = dp.delivery_person_id
GROUP BY dp.name
ORDER BY total_deliveries DESC;''')

headers = ["Delivery Person Name", "Total Deliveries"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```

```python
#15, Find the average delivery time per restaurant:

results = db_obj.fetch_all(''' SELECT r.name AS restaurant_name, AVG(d.delivery_time) AS avg_delivery_time
FROM delivery_table_data d
JOIN manages_order o ON d.order_id = o.order_id
JOIN manages_restaurant_infos r ON o.restaurant_id = r.restaurant_id
GROUP BY r.name
ORDER BY avg_delivery_time ASC;''')

headers = ["Restaurant Name", "Average Delivery Time"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```

```python
#16,Get the delivery persons who have completed more than 100 deliveries and have an average rating above 4.5:

results = db_obj.fetch_all('''SELECT name, total_deliveries, average_rating
FROM delivery_person_table
WHERE total_deliveries > 100 AND average_rating > 4.5;''')

headers = ["Name", "Total Deliveries", "Average Rating"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```

```python
#17, Find all orders that took longer than the estimated delivery time:

results = db_obj.fetch_all(''' SELECT d.order_id, d.delivery_time, d.estimated_time, d.delivery_status
FROM delivery_table_data d
WHERE d.delivery_time > d.estimated_time;''')

headers = ["Order ID", "Delivery Time", "Estimated Time", "Delivery Status"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```

```python
#18,Get the total number of orders per restaurant in descending order:

results = db_obj.fetch_all('''SELECT r.name AS restaurant_name, COUNT(o.order_id) AS total_orders
FROM manages_order o
JOIN manages_restaurant_infos r ON o.restaurant_id = r.restaurant_id
GROUP BY r.name
ORDER BY total_orders DESC;''')

headers = ["Restaurant Name", "Total Orders"]

print(tabulate(results, headers=headers, tablefmt="grid"))
```
